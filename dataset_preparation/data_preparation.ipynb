```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
data = pd.read_csv('datasets/features_datasets.csv')
data
```

```python
data.info()
```

```python
numeric_cols = ['aqi', 'co', 'no2', 'o3', 'pm10', 'pm25', 'so2', 'hour']
categorical_cols = ['city_name']
numeric_cols,categorical_cols
```

```python
encoder = OneHotEncoder(sparse_output=False)
encoder.fit(data[categorical_cols])

with open("train_test_data/encoder.pkl",'wb') as file:
    pickle.dump(encoder,file)
```

```python
data[categorical_cols].shape
```

```python
encoded_cities = encoder.get_feature_names_out(categorical_cols)
data[encoded_cities] = encoder.transform(data[categorical_cols])
data
```

```python
data = data.drop(['city_name'],axis=1)
data
```

```python
# splitting data into training and testing data
x = data.drop(['aqi'],axis=1)
y = data['aqi']
x.columns
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,shuffle=True,test_size=0.2)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.to_csv('train_test_data/x_train.csv',index=False)
x_test.to_csv("train_test_data/x_test.csv",index=False)
y_train.to_csv("train_test_data/y_train.csv",index=False)
y_test.to_csv("train_test_data/y_test.csv",index=False)
```
